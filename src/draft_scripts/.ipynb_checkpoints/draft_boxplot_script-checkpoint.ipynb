```r
library(readr)
library(tidyverse)
```

```r
data = read_csv('../../data/clean_data.csv')

sub1 = 'ABORIGINAL'
sub2 = 'NON ABORIGINAL'
fig_numeracy_path = '../../img/fig_test.png'
fig_reading_path = 
fig_writing_path = 

filtered_data = data %>%
  filter(sub_population == sub1 | sub_population == sub2)
```

```r
head(filtered_data)
```

```r
plot <- ggplot(filtered_data, aes(x = fsa_skill_code, y = score)) +
    geom_boxplot(aes(fill = sub_population)) + facet_grid(. ~ sub_population)

#plot
```

```r
plot <- ggplot(filtered_data, aes(x = sub_population, y = score)) +
    geom_boxplot(aes(fill = fsa_skill_code)) + facet_grid(. ~ sub_population)

#plot
```

```r
filtered_data_numeracy <- filtered_data %>%
    filter(fsa_skill_code == 'Numeracy')

filtered_data_reading <- filtered_data %>%
    filter(fsa_skill_code == 'Reading')

filtered_data_writing <- filtered_data %>%
    filter(fsa_skill_code == 'Writing')

boxplot_numeracy <- ggplot(filtered_data_numeracy, aes(x = sub_population, y = score)) +
    geom_boxplot(width = 0.7 , alpha=0.9 , size=0.3, colour="black") +
    labs(x = 'Sub Group',
         y = 'Average Score',
        title = 'FSA Numeracy Test Scores (2007/08 - 2018/19)') +
    stat_summary(fun.y = mean,
                geom = 'point',
                aes(shape = 'mean'),
                color = 'blue',
                size = 3) +
    scale_shape_manual('', values = c('mean' = 'triangle')) +
    theme_bw()
    #ggsave('../../img/fig_test.png', width = 8, height = 10)

boxplot_reading <- ggplot(filtered_data_reading, aes(x = sub_population, y = score)) +
    geom_boxplot(width = 0.7 , alpha=0.9 , size=0.3, colour="black") +
    labs(x = 'Sub Group',
         y = 'Average Score',
        title = 'FSA Reading Test Scores (2007/08 - 2018/19)') +
    stat_summary(fun.y = mean,
                geom = 'point',
                aes(shape = 'mean'),
                color = 'blue',
                size = 3) +
    scale_shape_manual('', values = c('mean' = 'triangle')) +
    theme_bw()
    #ggsave('../../img/fig_test.png', width = 8, height = 10)

boxplot_writing <- ggplot(filtered_data_writing, aes(x = sub_population, y = score)) +
    geom_boxplot(width = 0.7 , alpha=0.9 , size=0.3, colour="black") +
    labs(x = 'Sub Group',
         y = 'Average Score',
        title = 'FSA Writing Test Scores (2007/08 - 2018/19)') +
    stat_summary(fun.y = mean,
                geom = 'point',
                aes(shape = 'mean'),
                color = 'blue',
                size = 3) +
    scale_shape_manual('', values = c('mean' = 'triangle')) +
    theme_bw()
    #ggsave('../../img/fig_test.png', width = 8, height = 10)

boxplot_writing
```
